```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# quadratic eqn used  y=0.5x^2+1.5x+2+outliers
x=6* np.random.rand(100,1) - 3

y= 0.5* x**2 + 1.5*x + 2 + np.random.rand(100,1)
plt.scatter(x,y,color='r')
plt.xlabel('x dataset')
plt.ylabel('y dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# implemnet first simple linear regresiion
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
sscore=r2_score(y_test,regression_1.predict(x_test))
print(sscore)
```

```python
plt.plot(x_train,regression_1.predict(x_train),color="r")
plt.scatter(x_train,y_train)
plt.xlabel("x datset")
plt.ylabel("y")
```

```python
from sklearn.pipeline import Pipeline
```

```python


def poly_regression(degree):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)

    # Pipeline: polynomial expansion + linear regression
    model = Pipeline([
        ("poly_features", PolynomialFeatures(degree=degree, include_bias=True)),
        ("lin_reg", LinearRegression())
    ])

    # Fit model
    model.fit(x_train, y_train)

    # Predictions
    y_pred_new = model.predict(X_new)
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # R2 scores
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)

    # Plot prediction curve
    plt.figure(figsize=(8, 6))
    plt.plot(X_new, y_pred_new, 'r', label=f"Degree {degree} (Curve)", linewidth=2)

    # Plot train/test points with labels
    plt.scatter(x_train, y_train, color="blue", label="Train Data", s=40)
    plt.scatter(x_test, y_test, color="green", label="Test Data", s=40)

    # Text annotations inside plot
    plt.text(-3.8, 9.4, f"Train R² = {r2_train:.3f}", fontsize=10, color='blue')
    plt.text(-3.8, 8.7, f"Test R²  = {r2_test:.3f}", fontsize=10, color='green')

    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title(f"Polynomial Regression (Degree {degree})")
    plt.axis([-4, 4, 0, 10])
    plt.grid(True)
    plt.show()

    return r2_train, r2_test
```

```python
poly_regression(2)
```
